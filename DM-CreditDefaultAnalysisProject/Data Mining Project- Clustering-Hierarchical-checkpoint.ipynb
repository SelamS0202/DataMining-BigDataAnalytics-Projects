```python
def fancy_dendrogram(*args, **kwargs):
    max_d = kwargs.pop('max_d', None)
    if max_d and 'color_threshold' not in kwargs:
        kwargs['color_threshold'] = max_d
    annotate_above = kwargs.pop('annotate_above', 0)

    ddata = dendrogram(*args, **kwargs)

    if not kwargs.get('no_plot', False):
        plt.title('Hierarchical Clustering Dendrogram (Single link,Euclidean, Z-Score)')
        plt.xlabel('Cluster Index')
        plt.ylabel('distance')
        for i, d, c in zip(ddata['icoord'], ddata['dcoord'], ddata['color_list']):
            x = 0.5 * sum(i[1:3])
            y = d[1]
            if y > annotate_above:
                plt.plot(x, y, 'o', c=c)
                #plt.annotate("%.3g" % y, (x, y), xytext=(0, -5),
                             #textcoords='offset points',
                             #va='top', ha='center')
        if max_d:
            plt.axhline(y=max_d, c='k')
    return ddata
```

```python
def scatterPlot(dataset, first, second, column):
    column_values = sorted(dataset[column].unique())
    colors = ['r','g','B','b', 'c', 'm', 'y', 'k', 'w']
    plt.figure(figsize=(10,8))
    i = 0
    for values in column_values:
        plt.scatter(dataset[first][dataset[column] == values],dataset[second][dataset[column] == values], alpha=1,s=10,c=colors[i])
        i += 1

    plt.title('%s to %s Scatter Plot' % (first, second), fontsize=15)
    plt.xlabel(first,fontsize=15)
    plt.ylabel(second, fontsize=15)
    plt.legend(tuple(column_values), loc='best')
    ax = plt.gca()
    ax.tick_params(axis = 'both', which = 'major', labelsize = 15)
    plt.show()
```

```python
range_n_clusters = [2,3,4,5,6]
Silh = {}
#Silh = []
print('Method : Single Link\t\t  Metric : Euclidean' )
for n_clusters in range_n_clusters:
    Sclusters = fcluster(Single_link, n_clusters, criterion='maxclust')
    Silhouette = silhouette_score(x_Minmax, Sclusters)
    print (" K= ", n_clusters ,"The silhouette_score is :", Silhouette)
    Silh[n_clusters] = Silhouette
plt.plot(Silh.keys(), Silh.values())
plt.xlabel('Number of Clusters')
plt.title(' Method : Single Link,  Metric : Euclidean ')
plt.xticks(range_n_clusters)
plt.ylabel('silhouette')
plt.show()   
```

```python
def report(results, n_top=3):
    for i in range(1, n_top + 1):
        candidates = np.flatnonzero(results['rank_test_score'] == i)
        for candidate in candidates:
            print("Model with rank: {0}".format(i))
            print("Mean validation score: {0:.3f} (std: {1:.3f})".format(
                  results['mean_test_score'][candidate],
                  results['std_test_score'][candidate]))
            print("Parameters: {0}".format(results['params'][candidate]))
            print("")
```
